## Hill Climbing

```python
def in_board(state):
    
    k = len(state)
    
    for n in state:
        if n < 0 or n > k-1:
            return False 
        
    return True


def moves(state):
    
    k = len(state)
    
    row = np.random.randint(k)
        
    move = np.random.randint(-state[row],k-state[row])
    
    Newstate = state.copy()
    
    Newstate[row] = state[row] + move
        
    if heuristics(Newstate) == 0:

        print(Newstate)

        return Newstate

    else:

        if heuristics(Newstate) < heuristics(state):

            print(Newstate,heuristics(Newstate))

            moves(Newstate)

        else:
            print('repick')
            moves(state)
                
```

## Simulated annealing

```python
import math
import numpy as np

def in_board(state):
    
    k = len(state)
    
    for n in state:
        if n < 0 or n > k-1:
            return False 
        
    return True

Round = -100

def moves(state):
        
    global Round
    
    k = len(state)
    
    row = np.random.randint(k)
        
    move = np.random.randint(-state[row],k-state[row])
    
    Newstate = state.copy()
    
    Newstate[row] = state[row] + move
        
    if heuristics(Newstate) == 0:

        print(Newstate)

        return Newstate

    else:

        Choice = np.random.choice([0,1],p=[math.exp(-Round/100)/(1+math.exp(-Round/100)),1/(1+math.exp(-Round/100))])

        if Choice == 1:

            if heuristics(Newstate) < heuristics(state):

                print(Newstate,heuristics(Newstate),"Round=",Round,"Choice=",Choice)

                Round += 1

                moves(Newstate)

            else:

                print(Newstate,heuristics(Newstate),"Round=",Round,"Choice=",Choice)

                Round += 1

                moves(state)

        else:

            Round += 1

            print(Newstate,heuristics(Newstate),"Round=",Round,"Choice=",Choice)

            moves(Newstate)
                
```
