```python
def heuristics(state):
    
    count = 0
    
    for i in range(len(state)):
        
        for j in range(len(state)):
            
            if i != j:
            
                if state[i] == state[j]:

                    count += 1

                if abs(state[i] - state[j]) == abs(i-j):

                    count += 1
                
    return count
```

## Hill Climbing

```python


def in_board(state):
    
    k = len(state)
    
    for n in state:
        if n < 0 or n > k-1:
            return False 
        
    return True


def moves(state):
    
    k = len(state)
        
    move = np.random.randint(-4,5)
    
    row = np.random.randint(5)
    
    Newstate = state.copy()
    
    Newstate[row] = state[row] + move
    
    if in_board(Newstate):
        
        if heuristics(Newstate) == 0:
            
            print(Newstate)
            
            return Newstate
        
        else:
            
            if heuristics(Newstate) < heuristics(state):
                
                print(Newstate,heuristics(Newstate))

                moves(Newstate)

            else:
                moves(state)
                
    else:
        moves(state)
```

## Simulated annealing

```python
import math
import numpy as np

def in_board(state):
    
    k = len(state)
    
    for n in state:
        if n < 0 or n > k-1:
            return False 
        
    return True

Round = -11

def moves(state):
        
    global Round
    
    k = len(state)
        
    move = np.random.randint(-4,5)
    
    row = np.random.randint(5)
    
    Newstate = state.copy()
    
    Newstate[row] = state[row] + move
    
    if in_board(Newstate):
        
        if heuristics(Newstate) == 0:
            
            print(Newstate)
            
            return Newstate
        
        else:
            
            Choice = np.random.choice([0,1],p=[math.exp(-Round)/(1+math.exp(-Round)),1/(1+math.exp(-Round))])
            
            if Choice == 1:

                if heuristics(Newstate) < heuristics(state):

                    print(Newstate,heuristics(Newstate),"Round=",Round,"Choice=",Choice)
                    
                    Round += 1
                    
                    moves(Newstate)

                else:
                    
                    Round += 1
                    
                    moves(state)
                    
            else:
                
                Round += 1
                
                print(Newstate,heuristics(Newstate),"Round=",Round,"Choice=",Choice)
                
                moves(Newstate)
                
    else:
        moves(state)
```
