```python
import math
from IPython.display import clear_output

gameState1 = [[' ','X',' ','X',' ','X',' ','X'],
              ['X',' ','X',' ','X',' ','X',' '],
              [' ','X',' ','X',' ','X',' ','X'],
              [' ',' ',' ',' ',' ',' ',' ',' '],
              [' ',' ',' ',' ',' ',' ',' ',' '],
              ['O',' ','O',' ','O',' ','O',' '],       
              [' ','O',' ','O',' ','O',' ','O'],
              ['O',' ','O',' ','O',' ','O',' ']]

gameState2 = [[' ','X',' ',' ',' ',' ',' ',' '],
              [' ',' ',' ',' ','X',' ',' ',' '],
              [' ',' ',' ','O',' ',' ',' ',' '],
              [' ',' ',' ',' ',' ',' ',' ',' '],
              [' ',' ',' ',' ',' ',' ',' ',' '],
              ['O',' ',' ',' ','O',' ','O',' '],       
              [' ','O',' ','O',' ','O',' ','O'],
              ['O',' ','O',' ','O',' ','O',' ']]

gameState3 =  [ [' ', ' ', ' ', 'Q', ' ', ' ', ' ', 'X'],
                [' ', ' ', ' ', ' ', ' ', ' ', 'X', ' '],
                [' ', 'X', ' ', 'X', ' ', ' ', ' ', 'X'],
                ['X', ' ', 'X', ' ', 'X', ' ', 'X', ' '],
                [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
                ['O', ' ', ' ', ' ', 'O', ' ', 'O', ' '],
                [' ', 'O', ' ', 'O', ' ', 'O', ' ', ' '],
                [' ', ' ', 'O', ' ', ' ', ' ', 'O', ' ']]

global draw_count
draw_count = 40 
PlayerTurn = 'O'
maxDepth = 2

def TestWhoWin(cur):
    countX,countO,checkernum_X,checkernum_O,checkernum_K,checkernum_Q = degreecheck(cur)
    global draw_count
    if PlayerTurn == 'O' and countO == 0:
        print('X wins')
        return 'X wins'
    
    if PlayerTurn == 'X' and countX == 0:
        print('O wins')
        return 'O wins'
    
    if checkernum_O + checkernum_Q == 0:
        print('X wins')
        return 'X wins'
    if checkernum_X + checkernum_K == 0:
        print('O wins')
        return 'O wins'
   
    if draw_count == 0:
        print('Draw!')
        return 'Draw'
    
    return False

def KingMe(cur):
    new_cur = cur.copy()
    for i in range(len(new_cur[0])):
        if new_cur[0][i] == 'O':
            new_cur[0][i] = 'Q'
    for j in range(len(new_cur[7])):
        if new_cur[7][j] == 'X':
            new_cur[7][j] = 'K'
    return new_cur

def changePlayerTurn():
    global PlayerTurn
    if PlayerTurn == 'X':
        PlayerTurn = 'O'
    else:
        PlayerTurn = 'X'
        
        
def printGameBoard(cur,draw_count):
    row_number = 0
# We have to clear the previous output
    #clear_output()
# Some instructions to help our players
    print("Welcome to our checkers game!")
    print("Please use the row and column to select your checker and while placing your checker.")
    print("An X will change to a K and a O will turn to Q to represent a 'king' checker.")
    print("If you have made an incorrect selection you can escape using the 9 button.")
    print("Good luck!")
# We check after the player has moved if any checkers should be updated to a King
    cur = KingMe(cur)
# This helps players select row/column
    for i in cur:
        print(row_number,i)
        row_number += 1
    print('    0  ',' 1  ',' 2  ',' 3  ',' 4  ',' 5 ','  6  ',' 7')
    print("A player must make a leap in:", draw_count, "moves. Otherwise there will be a draw.")
    
def degreecheck(cur):
# This function is to understand if a player has any degrees of freedom. 
# This is used in our win condition as a player could have a checker on the board but not
# be able to move it, this indicates there are 0 degrees of freedom and is considered a loss. 
    countO = 0
    countX = 0
    checkernum_O = 0
    checkernum_X = 0
    checkernum_Q = 0
    checkernum_K = 0
    for i in range(8):
        for j in range(8):
            if cur[i][j] == 'X':
                checkernum_X += 1
                if i+1<=7 and j+1<=7 and cur[i+1][j+1] == ' ':
                    countX += 1
                if i+1<=7 and j-1>=0 and cur[i+1][j-1] == ' ':
                    countX += 1   
            elif cur[i][j] == 'O':
                checkernum_O += 1
                if i-1 >=0 and j+1 <=7 and cur[i-1][j+1] == ' ':
                    countO += 1
                if i-1 >=0 and j-1>=0 and cur[i-1][j-1] == ' ':
                    countO += 1
            elif cur[i][j] == 'K':
                checkernum_K += 1
                if 0<= i+1 <=7 and 0<= j+1 <=7 and cur[i+1][j+1] == ' ':
                    countX += 1    
                if 0<= i+1 <=7 and 0<= j-1 <=7 and cur[i+1][j-1] == ' ':
                    countX += 1
                if 0<= i-1 <=7 and 0<= j+1 <=7 and cur[i-1][j+1] == ' ':
                    countX += 1
                if 0<= i-1 <=7 and 0<= j-1 <=7 and cur[i-1][j-1] == ' ':
                    countX += 1
            elif cur[i][j] == 'Q':
                checkernum_Q += 1 
                if 0<= i+1 <=7 and 0<= j+1 <=7 and cur[i+1][j+1] == ' ':
                    countO += 1
                if 0<= i+1 <=7 and 0<= j-1 <=7 and cur[i+1][j-1] == ' ':
                    countO += 1
                if 0<= i-1 <=7 and 0<= j+1 <=7 and cur[i-1][j+1] == ' ':
                    countO += 1
                if 0<= i-1 <=7 and 0<= j-1 <=7 and cur[i-1][j-1] == ' ':
                    countO += 1    
    return countX, countO, checkernum_X, checkernum_O, checkernum_K, checkernum_Q
            
def degree(cur):
# We can use our degrees of freedome concept to understand the utility of nodes
# A checker is worth 5 and a king/queen is worth 10.
    countX, countO, checkernum_X, checkernum_O, checkernum_K, checkernum_Q = degreecheck(cur)
    return (countX - countO) + 5*(checkernum_X - checkernum_O) + 10*(checkernum_K - checkernum_Q)
    
def gotwoways(cur,i,j):
# This function makes a list of possible moves for each checker on the board
    jumpState = []
# All moves for X
    if cur[i][j] == 'X':
        if 0 <= i + 2 <8 and 0<=j + 2 < 8 and cur[i + 2][j + 2] == ' ' and cur[i+1][j+1] in ('O','Q'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i+2][j+2] = 'X'
            nextState[i][j] = ' '
            nextState[i+1][j+1] = ' '
            gotwoways(nextState,i+2,j+2)
        if 0 <= i + 2 < 8 and 0 <=j - 2 < 8 and cur[i + 2][j - 2] == ' ' and cur[i+1][j-1] in ('O','Q'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i+2][j-2] = 'X'
            nextState[i][j] = ' '
            nextState[i+1][j-1] = ' '
            gotwoways(nextState,i+2,j-2)
        else:
            jumpState.append(cur)
    elif cur[i][j] == 'K':
# All moves for K
        if 0 <= i + 2 <8 and 0<=j + 2 < 8 and cur[i + 2][j + 2] == ' ' and cur[i+1][j+1] in ('O','Q'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i+2][j+2] = 'K'
            nextState[i][j] = ' '
            nextState[i+1][j+1] = ' '
            gotwoways(nextState,i+2,j+2)
        if 0 <= i + 2 < 8 and 0 <=j - 2 < 8 and cur[i + 2][j - 2] == ' ' and cur[i+1][j-1] in ('O','Q'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i+2][j-2] = 'K'
            nextState[i][j] = ' '
            nextState[i+1][j-1] = ' '
            gotwoways(nextState,i+2,j-2)
        if 0 <= i - 2 <8 and 0<=j + 2 < 8 and cur[i - 2][j + 2] == ' ' and cur[i-1][j+1] in ('O','Q'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i-2][j+2] = 'K'
            nextState[i][j] = ' '
            nextState[i-1][j+1] = ' '
            gotwoways(nextState,i-2,j+2)
        if 0 <= i - 2 < 8 and 0 <=j - 2 < 8 and cur[i - 2][j - 2] == ' ' and cur[i-1][j-1] in ('O','Q'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i-2][j-2] = 'K'
            nextState[i][j] = ' '
            nextState[i-1][j-1] = ' '
            gotwoways(nextState,i-2,j-2)
        else:
            jumpState.append(cur)           
    elif cur[i][j] == 'O':
# All moves for O
        if 0 <= i - 2 <8 and 0<=j + 2 < 8 and cur[i - 2][j + 2] == ' ' and cur[i-1][j+1] in ('K','X'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i-2][j+2] = 'O'
            nextState[i][j] = ' '
            nextState[i-1][j+1] = ' '
            gotwoways(nextState,i-2,j+2)
        if 0 <= i - 2 < 8 and 0 <=j - 2 < 8 and cur[i - 2][j - 2] == ' ' and cur[i-1][j-1] in ('K','X'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i-2][j-2] = 'O'
            nextState[i][j] = ' '
            nextState[i-1][j-1] = ' '
            gotwoways(nextState,i-2,j-2)    
        else:
            jumpState.append(cur)              
    elif cur[i][j] == 'Q':
# All moves for Q
        if 0 <= i - 2 <8 and 0<=j + 2 < 8 and cur[i - 2][j + 2] == ' ' and cur[i-1][j+1] in ('K','X'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i-2][j+2] = 'O'
            nextState[i][j] = ' '
            nextState[i-1][j+1] = ' '
            gotwoways(nextState,i-2,j+2)
        if 0 <= i - 2 < 8 and 0 <=j - 2 < 8 and cur[i - 2][j - 2] == ' ' and cur[i-1][j-1] in ('K','X'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i-2][j-2] = 'O'
            nextState[i][j] = ' '
            nextState[i-1][j-1] = ' '
            gotwoways(nextState,i-2,j-2)
        if 0 <= i + 2 <8 and 0<=j + 2 < 8 and cur[i + 2][j + 2] == ' ' and cur[i+1][j+1] in ('K','X'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i-2][j+2] = 'O'
            nextState[i][j] = ' '
            nextState[i-1][j+1] = ' '
            gotwoways(nextState,i-2,j+2)
        if 0 <= i + 2 < 8 and 0 <=j - 2 < 8 and cur[i + 2][j - 2] == ' ' and cur[i+1][j-1] in ('K','X'):
            nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
            nextState[i-2][j-2] = 'O'
            nextState[i][j] = ' '
            nextState[i-1][j-1] = ' '
            gotwoways(nextState,i-2,j-2)
        else:
            jumpState.append(cur)       
    return jumpState
# The main function
def MinimaxDecision (cur):# check the Max value for all possible actions
 
    maxVal = -math.inf
    alpha = -math.inf
    beta = math.inf
    global state
    global draw_count
    draw_count -= 1
    state = []
    jumpState = []
    for i in range(8):
        for j in range(8):  
# check if we can make a leap (attacking move)           
            if ifleap(cur, i, j) and cur[i][j] in ('K','X'):
                print('ifleap')
                Tempo = gotwoways(cur,i,j)
                jumpState.extend(Tempo)
                print(jumpState)
    if jumpState:
        draw_count = 40
        for a in jumpState:
            val = MinValue(a, alpha, beta, maxDepth)
            if (val > maxVal):
                maxVal = val
                state = [[a[ti][tj] for tj in range(8)] for ti in range(8)]
            if (val > alpha):
                alpha = val
        return state
# We can use our gotwoways function to understand the utility of nodes          
    for i in range(8):
        for j in range(8):  
# check jump (non-attacking move)
            if cur[i][j] in ('K','X') and ((0 <= i+1 < 8 and 0 <= j-1 < 8 and (cur[i+1][j-1] == ' ')) or (0 <= i+1 < 8 and 0 <=j+1 < 8 and cur[i+1][j+1] == ' ')):
# finding our Alpha and V for X's moves
                if (0 <= i+1 < 8 and 0 <= j-1 < 8 and cur[i+1][j-1] == ' '):
                    
                    nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
                    nextState[i+1][j-1] = nextState[i][j]
                    nextState[i][j] = ' '
                    val = MinValue(nextState, alpha, beta, maxDepth)
                    
                    if (val > maxVal):
                        maxVal = val
                        state = [[nextState[ti][tj] for tj in range(8)] for ti in range(8)]
                        
                    if (val > alpha):
                        alpha = val
                
                if (0 <= i+1 < 8 and 0 <=j+1 < 8 and cur[i+1][j+1] == ' '): 
                    
                    nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
                    nextState[i+1][j+1] = nextState[i][j]
                    nextState[i][j] = ' '
                    val = MinValue(nextState, alpha, beta, maxDepth) 
                    

                    if (val > maxVal):
                        maxVal = val
                        state = [[nextState[ti][tj] for tj in range(8)] for ti in range(8)]
                        
                    if (val > alpha):
                        alpha = val
                        
    return state

def MaxValue (cur, alpha, beta, depth):
  # we have 3 terminal states: X O Draw
    if TestWhoWin(cur) == 'X wins':
        return 100
    if TestWhoWin(cur) == 'O wins':
        return -100
    if TestWhoWin(cur) == 'Draw':
        return 0
    if (depth == 0):
        return degree(cur)
  # check the Max value for all possible actions
    maxVal = -math.inf
    jumpState = []
 # We check if there is possible leap for all checkers first as it is mandatory to leap if there is an available leap
    for i in range(8): 
        for j in range(8):          
            if ifleap(cur, i, j) and cur[i][j] in ('K','X'):                    
                jumpState = gotwoways(cur,i,j)
                            # This will reset draw count to 40 if an opposing unit has been jumped
    if jumpState: 
            # In case there is no available leap so that it is an empty list
        for a in jumpState:                         
            val = MinValue(a, alpha, beta, depth-1)  
            if (val > maxVal):
                maxVal = val
            if (val >= beta):
                return val
            if (val > alpha):
                alpha = val
        return maxVal
        
    for i in range(8):                      # Normal step (Non-attacking move)
        for j in range(8):
                
            if (cur[i][j] in ('K','X')) and ((0 <= i+1 < 8 and 0 <= j-1 < 8 and cur[i+1][j-1] == ' ') or (0 <= i+1 < 8 and 0 <=j+1 < 8 and cur[i+1][j+1] == ' ')):
                
                if (0 <= i+1 < 8 and 0 <= j-1 < 8 and cur[i+1][j-1] == ' '):
                    
                    nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
                    nextState[i+1][j-1] = cur[i][j]
                    nextState[i][j] = ' '
                    val = MinValue(nextState, alpha, beta, depth-1)

                    if (val > maxVal):
                        maxVal = val
                    if (val >= beta):
                        return val
                    if (val > alpha):
                        alpha = val
            
                if (0 <= i+1 < 8 and 0 <=j+1 < 8 and cur[i+1][j+1] == ' '):
                
                    nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
                    nextState[i+1][j+1] = cur[i][j]
                    nextState[i][j] = ' '

                    val = MinValue(nextState, alpha, beta, depth-1)
                    if (val > maxVal):
                        maxVal = val
                    if (val >= beta):
                        return val
                    if (val > alpha):
                        alpha = val
            
    return maxVal

def MinValue (cur, alpha, beta, depth):
    if TestWhoWin(cur) == 'X wins':
        return 100
    if TestWhoWin(cur) == 'O wins':
        return -100
    if TestWhoWin(cur) == 'Draw':
        return 0
    if (depth == 0):
        return degree(cur)
    minVal = math.inf 
    jumpState = []

    for i in range(8):
        for j in range(8):
            if ifleap(cur, i, j) and cur[i][j] in ('Q','O'):                                                  
                jumpState = gotwoways(cur,i,j)
        
    if jumpState:
        for a in jumpState:
            val = MaxValue(a, alpha, beta, depth-1)
            if (val < minVal):
                minVal = val
            if (val <= alpha):
                return val
            if (val < beta):
                beta = val
        return minVal
        
    for i in range(8):
        for j in range(8):
 
            if cur[i][j] in ('Q','O') and ((0 <= i-1 < 8 and 0 <=j-1 < 8 and cur[i-1][j-1] == ' ') or (0 <= i-1 < 8 and 0 <=j+1 < 8 and cur[i-1][j+1] == ' ')):
                                       
                if (0 <= i-1 < 8 and 0 <=j-1 < 8 and cur[i-1][j-1] == ' '):
        
                    nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
                    nextState[i-1][j-1] = cur[i][j]
                    nextState[i][j] = ' '
                    val = MaxValue(nextState, alpha, beta, depth-1)
                                           
                    if (val < minVal):
                        minVal = val
                    if (val <= alpha):
                        return val
                    if (val < beta):
                        beta = val
            
                if (0 <= i-1 < 8 and 0 <=j+1 < 8 and cur[i-1][j+1] == ' '):
                    
                    nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
                    nextState[i-1][j+1] = cur[i][j]
                    nextState[i][j] = ' '
                    val = MaxValue(nextState, alpha, beta, depth-1)
                    if (val < minVal):
                        minVal = val
                    if (val <= alpha):
                        return val
                    if (val < beta):
                        beta = val
   
    return minVal

def ifleap(gameState, check_x, check_y):
    # if O - X/K
    if gameState[check_x][check_y] in ('Q','O'):
        if ( 0<=check_x-2<=7 and 0<=check_y+2<=7 and gameState[check_x-1][check_y+1] in ('X','K') and gameState[check_x-2][check_y+2] == ' ') or (0<=check_x-2<=7 and 0<=check_y-2<=7 and gameState[check_x-1][check_y-1] in ('X','K') and gameState[check_x-2][check_y-2] == ' ') :
            return True
    # if Q - X/K
    #elif gameState[check_x][check_y] == 'Q': 
    #    if ( 0<=check_x-2<=7 and 0<=check_y+2<=7 and gameState[check_x-1][check_y+1] in ('X','K') and gameState[check_x-2][check_y+2] == ' ') or (0<=check_x-2<=7 and 0<=check_y-2<=7 and gameState[check_x-1][check_y-1] in ('X','K') and gameState[check_x-2][check_y-2] == ' ') or ( 0<=check_x+2<=7 and 0<=check_y-2<=7 and gameState[check_x+1][check_y-1] in ('X','K') and gameState[check_x+2][check_y-2] == ' ') or (0<=check_x+2<=7 and 0<=check_y+2<=7 and gameState[check_x+1][check_y+1] == ('X' in 'K') and gameState[check_x+2][check_y+2] == ' '):
     #       return True    
    # if X - O/Q
    elif gameState[check_x][check_y] in ('K','X'):
        if ( 0<=check_x+2<=7 and 0<=check_y-2<=7 and gameState[check_x+1][check_y-1] in ('O','Q') and gameState[check_x+2][check_y-2] == ' ') or (0<=check_x+2<=7 and 0<=check_y+2<=7 and gameState[check_x+1][check_y+1] in ('O','Q') and gameState[check_x+2][check_y+2] == ' '):
            return True
    # if K - O/Q
    #elif gameState[check_x][check_y] == 'K':
    #    if ( 0<=check_x-2<=7 and 0<=check_y+2<=7 and gameState[check_x-1][check_y+1] in ('O','Q') and gameState[check_x-2][check_y+2] == ' ') or (0<=check_x-2<=7 and 0<=check_y-2<=7 and gameState[check_x-1][check_y-1] in ('O','Q') and gameState[check_x-2][check_y-2] == ' ') or ( 0<=check_x+2<=7 and 0<=check_y-2<=7 and gameState[check_x+1][check_y-1] == ('O' or 'Q') and gameState[check_x+2][check_y-2] == ' ') or (0<=check_x+2<=7 and 0<=check_y+2<=7 and gameState[check_x+1][check_y+1] == ('O' or 'Q') and gameState[check_x+2][check_y+2] == ' '):
    #        return True
    return False
    
    
def input_info(gameState):
# This function handles the user's inputs. 
    check_x = int(input("Please provide the row number of the checker you would like to move: "))
    check_y = int(input("Please provide the column number of the checker you would like to move: "))
# This will handle errors until the user is provides a correct checker    
    while (check_x < 0 or check_x > 7 or check_y < 0 or check_y > 7 or gameState[check_x][check_y] not in ('O','Q')):
        check_x = int(input("Please provide a valid checker row number "))
        check_y = int(input("Please provide a valid checker column number "))
# We can give the user an escape by selecting 9 
    check_x_dir = int(input("Please provide the row number of where you would like to move: "))   
    if check_x_dir ==9:
        input_info(gameState)
    check_y_dir = int(input("Please provide the column number of where you would like to move: "))
    if check_y_dir ==9:
        input_info(gameState)
    while (gameState[check_x_dir][check_y_dir] != ' ') or ((abs(check_x_dir-check_x)!=1 or abs(check_y_dir-check_y)!=1) and ((abs(check_x_dir-check_x)!=2 or abs(check_y_dir-check_y)!=2))):
        check_x_dir = int(input("Please provide a valid row number "))
        if check_x_dir ==9:
            input_info(gameState)
        check_y_dir = int(input("Please provide  a valid column number "))
        if check_x_dir ==9:
            input_info(gameState)
    if ifleap(gameState,check_x,check_y):
        leap(gameState, check_x, check_y,check_x_dir, check_y_dir)
#We reset the draw_count after we have leap an enemy unit
        global draw_count
        draw_count = 40
    elif gameState[check_x_dir][check_y_dir] == ' ':
        draw_count -= 1
       # if gameState[check_x][check_y] == 'O':
        gameState[check_x_dir][check_y_dir] = gameState[check_x][check_y]
        gameState[check_x][check_y] = ' '
     #   else:
     #       gameState[check_x_dir][check_y_dir] = 'Q'
     #       gameState[check_x][check_y] = ' '            
        return gameState
    
        

def leap(gameState, check_x, check_y,check_x_dir, check_y_dir):  
# Checkers allows for a second leap if the checker has leaped another checker
    gameState[(int((check_x + check_x_dir)/2))][int((check_y+check_y_dir)/2)] = ' '
    if gameState[check_x][check_y] in ('O', 'Q'):
        gameState[check_x_dir][check_y_dir] = gameState[check_x][check_y]
        gameState[check_x][check_y] = ' '
    printGameBoard(gameState,draw_count)
# Checkers requires leaps to be taken therefore, if you can leap you must leap! 
    if ifleap(gameState, check_x_dir, check_y_dir):
        check_x = check_x_dir 
        check_y = check_y_dir 
        check_x_dir = int(input("Please provide the row number of where you would like to move: "))
        check_y_dir = int(input("Please provide the column number of where you would like to move: "))
        while ((check_x_dir < 0 or check_x_dir > 7 or check_y_dir < 0 or check_y_dir > 7) or gameState[check_x_dir][check_y_dir] != ' '):
            check_x_dir = int(input("Please provide a valid row number."))
            check_y_dir = int(input("Please provide a valid column number.")) 
# This will loop until the player is not leaping any more units
        leap(gameState, check_x, check_y, check_x_dir, check_y_dir)
    else: 
        return gameState 


```

```python
def monteCarloTreeSearch (cur):
    # first check if the cur state is already added to the stateEvalDict, if not create an empty place for it
    if (checkIfExistInEvalDict(cur) == False):
        addStateToStateEvalDict (cur)
    # keep running the loop and testing for 3 seconds
    start_time = time.time()
    oldTime = 11
    while (True):
        # print timer
        newTime = 10 - (int)(time.time() - start_time)
        if (oldTime != newTime):
            oldTime = newTime
            print(oldTime, end=" ")
        if ((time.time() - start_time) > AITime): # if difference time more than 3 seconds, break
            break
            
        # since we have time, execute the following parts
        # selection Part
        stateToExplore, PlayerID, endGame = selectionPart(cur)
        print(stateToExplore,endGame)
        
        if (endGame == False): # the game didn't end in the selection part, so let us expand            
            # Expansion Part
            addStateToStateEvalDict(stateToExplore)
            # add it to statesToBeUpdInEvalDict, so it can be updated in the BackPropagation part
            curStr = ''
            for j in stateToExplore:
                for i in j:
                    curStr += str(i)
            statesToBeUpdInEvalDict.append(curStr)
            returnedUtility = playOut(stateToExplore,PlayerID)
            #print(statesToBeUpdInEvalDict)

        else: # what if a terminal state was reached in the selection part
            # check who won, and return utility
            if PlayerID == 'X':
                returnedUtility = winningTerminalState
            elif PlayerID == 'O':
                returnedUtility = loosingTerminalState
            else:
                returnedUtility = drawTerminalState
        
        # Backpropagation part
        for eachState in statesToBeUpdInEvalDict:
            curEval = stateEvalDict[eachState]
            curEval[0] += returnedUtility
            curEval[1] += 1
            stateEvalDict[eachState] = curEval
             
    # when we stop, we choose the best action so far
    maxUCB = -math.inf
    state = []
    # save the eval and nbTimes of current or parent state
    currentStateEval = checkIfExistInEvalDict(cur)
    # check all actions
    for i in range(8):
        for j in range(8):  
# check if we can make a leap (attacking move)           
            if ifleap(cur, i, j) and cur[i][j] in ('K','X'):
                gotwoways(cur, i, j)
                jumpState = gotwoways(cur,i,j)
    
    if jumpState:
        for a in jumpState:
            nextState = [[a[ti][tj] for tj in range(8)] for ti in range(8)]
            nextStateEval = checkIfExistInEvalDict(nextState)
            
            # stateUCB = vi + C * sqrt ( ln (N) / ni)
            vi = nextStateEval[0] / nextStateEval[1]
            ln = math.log(currentStateEval[1])
            secondpart = ln / nextStateEval[1]
            secondpart = math.sqrt(secondpart)
            stateUCB = vi + learningRateC * secondpart
            
            if (stateUCB > maxUCB): # new max
                maxUCB = stateUCB
                state = [[nextState[ti][tj] for tj in range(8)] for ti in range(8)]
        return state
    
    for i in range(8):
        for j in range(8):  
# check jump (non-attacking move)
            if (cur[i][j] in ('K','X')) and ((0 <= i+1 < 8 and 0 <= j-1 < 8 and (cur[i+1][j-1] == ' ')) or (0 <= i+1 < 8 and 0 <=j+1 < 8 and cur[i+1][j+1] == ' ')):
# finding our Alpha and V for X's moves
                if (0 <= i+1 < 8 and 0 <= j-1 < 8 and cur[i+1][j-1] == ' '):
                    nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
                    nextState[i+1][j-1] = nextState[i][j]
                    nextState[i][j] = ' '
                    nextStateEval = checkIfExistInEvalDict(nextState)

                    # stateUCB = vi + C * sqrt ( ln (N) / ni)
                    vi = nextStateEval[0] / nextStateEval[1]
                    ln = math.log(currentStateEval[1])
                    secondpart = ln / nextStateEval[1]
                    secondpart = math.sqrt(secondpart)
                    stateUCB = vi + learningRateC * secondpart

                    if (stateUCB > maxUCB): # new max
                        maxUCB = stateUCB
                        state = [[nextState[ti][tj] for tj in range(8)] for ti in range(8)]

                if (0 <= i+1 < 8 and 0 <=j+1 < 8 and cur[i+1][j+1] == ' '): 

                    nextState = [[cur[ti][tj] for tj in range(8)] for ti in range(8)]
                    nextState[i+1][j+1] = nextState[i][j]
                    nextState[i][j] = ' '
                    nextStateEval = checkIfExistInEvalDict(nextState)

                    # stateUCB = vi + C * sqrt ( ln (N) / ni)
                    vi = nextStateEval[0] / nextStateEval[1]
                    ln = math.log(currentStateEval[1])
                    secondpart = ln / nextStateEval[1]
                    secondpart = math.sqrt(secondpart)
                    stateUCB = vi + learningRateC * secondpart

                    if (stateUCB > maxUCB): # new max
                        maxUCB = stateUCB
                        state = [[nextState[ti][tj] for tj in range(8)] for ti in range(8)]
                        
                        
    return state
```
